(c) Recode a numeric colum

<font size = "3">

- Import the "data_raw/races.csv" file
- Recode the "year" column into a new column "year_brackets" <br>
with the following categories

$\qquad$ ``` ["1950-1965","1966-1980","1980-onwards"] ```

$\qquad$ HINT: Use the "pd.cut()" command.

````python
# Write your own code

races = pd.read_csv("data_raw/races.csv")

bins_races = [1950, 1965, 1980, float("inf")]
labels_races = ["Between 1950 and 1965", "Between 1966 and 1980", "1980-onwards"]

races["year_brackets"] = pd.cut(races["year"], bins=bins_races, labels=labels_races, right=True)

# races

````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the maximum of "laps" by "driverId" <br> 
using ``` .groupby().agg() ```
- Sort the teams in descending order using of <br>
the maximum of laps using ``` .sort_values() ```

````python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create the figure and axes object
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the data for the cone
r = np.linspace(0, 0.5, 50)
theta = np.linspace(0, 2*np.pi, 50)
r, theta = np.meshgrid(r, theta)

x_cone = r * np.cos(theta)
y_cone = r * np.sin(theta)
z_cone = r

# Create the data for the sphere
u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi/2, 50)  # phi ranges from 0 to pi/2
x_sphere = 0.5 * np.sin(v) * np.cos(u)
y_sphere = 0.5 * np.sin(v) * np.sin(u)
z_sphere = 0.5 + 0.5 * np.cos(v)  # Shift the sphere up by 0.5

# Plot the cone
ax.plot_surface(x_cone, y_cone, z_cone, color='blue', alpha=0.5, label='Cone')

# Plot the sphere
ax.plot_surface(x_sphere, y_sphere, z_sphere, color='red', alpha=0.5, label='Sphere')

# Set the axis labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set the axis limits
ax.set_xlim([-0.5, 0.5])
ax.set_ylim([-0.5, 0.5])
ax.set_zlim([0, 1])

# Add a legend
ax.legend()

# Show the plot
plt.title('Visualization of the Solid')
plt.show()
````

````python
# Write your own code

sprint_results = pd.read_csv("data_raw/sprint_results.csv")

sprint_results_agg = (sprint_results.groupby(["driverId"])
                       .agg(max_laps = ('laps', 'max'))
                       .sort_values("max_laps", ascending=False))

# sprint_results_agg






````
