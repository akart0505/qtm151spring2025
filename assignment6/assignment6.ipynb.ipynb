- Read the dataset labelled features.csv.

if displacement $\leq$ 200.
- Count how many cars have displacement $> 200$.

To get full credit use either the apply or the map function <br>
(see Lectures 7 and 8)

````python
# Write your answer here:

data = pd.read_csv("/Users/ajaykartik/Documents/GitHub/qtm151spring2025/assignment6/data/features.csv")

displacement_greater_than_200 = lambda x: x > 200

data["displacement_check"] = data["displacement"].apply(displacement_greater_than_200)

cars_over_200 = data["displacement_check"].sum()

print(f"The number of cars that have a displacement > 200 is: {cars_over_200}")






````

(b) Subsetting Data

<font size = "3" >

- Read the dataset labelled features.csv.
- Define a global variable called ```threshold``` that takes the <br>
value 2588
- Subset the data to get one dataset with weight $>$ threshold and <br>
another with weight $\le$ 2588. Use the ```.query()``` function.
- Compute the mean acceleration for each of these datasets.<br>

Note: Review how to define a gloabl variable inside ```.query()```
 

````python
# Write your answer here:

data = pd.read_csv("/Users/ajaykartik/Documents/GitHub/qtm151spring2025/assignment6/data/features.csv")

threshold = 2588

above_threshold = data.query("weight > @threshold")
equal_below_threshold = data.query("weight <= @threshold")

mean_acceleration_above = above_threshold["acceleration"].mean()
mean_acceleration_below = equal_below_threshold["acceleration"].mean()

print(f"The mean acceleration for weight above {threshold} is: {mean_acceleration_above}")
print(f"The mean acceleration for weight equal to or below {threshold} is: {mean_acceleration_below}")


````

(d) Use queries + global variables

<font size = "3">

- Read the "fifa23_players_basic.csv" dataset
- Compute the mean of the column "Wage(in Euro)" <br>
and store it as a variable
- Subset the players whose value is above the mean value <br>
  using ".query()"


HINT: Remember that there is a special syntax for <br>
when the column names have spaces (see Lecture 10) <br>
Reference the column as ``` `Wage(in Euro)`    ``` inside .query()

````python
# Write your answer here:

fifa_data = pd.read_csv("/Users/ajaykartik/Documents/GitHub/qtm151spring2025/assignment6/data/fifa23_players_basic.csv")

mean_wage = fifa_data["Wage(in Euro)"].mean()

players_above_mean = fifa_data.query("`Wage(in Euro)` > @mean_wage")

print(f"The mean wage (in Euros) is: {mean_wage}")
print(f"The subset of players whose value is above the mean is: {players_above_mean}")

````

Read the "data/wdi_urban.csv" dataset.

  This contains data from 1980 and 2020.

- **Create a list with the unique values of "year"**

- **Create a plot with a ...**
    - Histogram of "prop_urbanpopulation" for each year
    - Set the histogram option `alpha = 0.5` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title


````python
# Write your answer here:

wdi_data = pd.read_csv("/Users/ajaykartik/Documents/GitHub/qtm151spring2025/assignment6/data/wdi_urban.csv")

years = wdi_data["year"].unique()

for year in years:
    yearly = wdi_data[wdi_data["year"] == year]
    plt.hist(yearly["prop_urbanpopulation"], alpha = 0.5)

plt.legend(labels = years, title="year")
plt.xlabel("Proportion of Urban Population")
plt.ylabel("Frequency")
plt.title("Urban Population Proportion Per Year ")
plt.show()








````
